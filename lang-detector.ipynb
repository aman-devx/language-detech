```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
data = pd.read_csv("https://raw.githubusercontent.com/TheiScale/YouTube-Video-Notes/refs/heads/main/Machine%20Learning%2BNLP%20Project%20Language%20Detection/language.csv")
print(data.head())
```

```python
x = np.array(data['Text'])
y = np.array(data['language'])
```

```python
cv = CountVectorizer()
X = cv.fit_transform(x)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=100)
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
user = input("Enter text to detect language: ")
data = cv.transform([user]).toarray()
output = model.predict(data)
print("Predicted language:", output)
```

```python
# Save the model using pickle
import pickle
pickle_out = open("lang_detect_model.pkl", "wb")
pickle.dump(model, pickle_out)
pickle_out.close()
```
